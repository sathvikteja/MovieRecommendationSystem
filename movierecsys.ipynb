```python
# i only want the genres by removing all other items
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
# we want to make "sam worthington" to "samworthington"
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# creating tags col by concatenating overview, genres, cast, crew, keywords
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

Problem Statement: "When a user queries a movie we have to return top 5 most similar movies to that movie" 

```python
import nltk
```

But the most frequent words can contain duplicated words like (action, actions)

So we do **Stemming**

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

Similarity between movies can be understood by the text similarity between tags column of that movie

So we convert these tags(words) to vectors so that it becomes easy to find the similarity between movies

We use **BOW** technique:

For **BOW** we are only considering our vocabulary as the top 5000 most occured words from the corpus (combination all movies tags) 

So that every movie tag vector dimension will be 5000

We will ignore StopWords

We are using **sklearn** library called **CountVectorizer** or all this

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
len(cv.get_feature_names_out())
```

We need to find the similarity (in terms of distance) of each movie with every other movies

We use Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

shape is (4806,4806)

```python
similarity=cosine_similarity(vectors.astype('float32'))
```

```python
similarity
```

```python
similarity.shape
```

```python
#similarity of first movie with all other movies
similarity[0]
```

```python
similarity[1]
```

```python
#if one movie is given we have to return the top 5 most similar movies
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
